Ques2 [R] Plot a histogram of mpg attribute using seaborn library. Use bin sizes from {default, 20,50, 100, 200}. Use KDE to plot the density graphs. Report the observations.

### Answer:

### Q4. [R] Prepare a bar plot for mpg vs displacement and group according to model year using median estimator. Add a legend at an appropriate location. Report your observations.
### Answer:



```python
plt.figure(figsize=(100,8))
ax=sns.barplot(x='mpg',y='displacement',data=df,hue='model_year',estimator=np.median)
ax.legend(loc='center left', title="model_year", bbox_to_anchor=(1, 0.5));
```

### Answer:
Since error estimator is median here and in above plot error is higher in initial values but as soon as value of mpg increases then value of displacement also decreases.Since there is difference between bars this shows that displacement is not equally distributed in mpg.

### Q8. [R] Prepare a box plot between model year and displacement and group according to origin. Discuss the observations.
### Solution:

```python
plt.figure(figsize=(20,15))
ax=sns.boxplot(x='model_year',y='displacement',data=df,hue='origin')
ax.legend(loc='center left', title="origin", bbox_to_anchor=(1, 0.5));
```

### Answer:
By above plots we can observe that in initial years difference in USA and other countries is high in displacement.But as soon as time passes this difference slightly decreases and in year(81-82) difference is minimum in displacement of USA and other countries.For Japan and Europe displace has a little difference.For USA error bar(95% confidence interval) is highly longer which shows that data displacement is spread widely.

### Q9. [R] Prepare a violin plot between model year and acceleration and group according to origin. Discuss the observations.

### Q10. [R] Prepare a pairplot with displacement and horsepower along x axis and acceleration along y axis, with kind=reg and cylinders as hue. Explain the observations obtained from this pairplot.
### Answer:


```python

sns.pairplot(x_vars=['displacement','horsepower'],y_vars=['acceleration'],data=df,kind='reg',hue='cylinders');
```

### Answer:
Since regression line denotes the relation between data in x-axis and data in y-axis that is if regression line is decreasing then both data are inversely porpotional to each other else both are directly porpotional to each other.We can also relate each cylinders to each other which depends on slope of regression line if slope is high then slope of other regression line then corresponding data will fastly increases in comparison of respective cylinder's other regression line.
