Plot these loss functions Gh(z), Gℓ(z) and Gsh(z)} \\ \text{ where z takes values on the real line [−∞, ∞]. Distinguish the loss
functions using different colors. Comment on the behavior of} \\ \text{ respective loss functions with respect to z}$

```python
def hinge_loss(arr):
  loss=[]
  for i in arr:
    loss.append(max(0,1-i))
  return loss  

```

```python
def log_loss(arr):
  loss=[]
  for i in arr:
    loss.append(np.log(1+np.exp(-i)))
  return loss  
```

```python
def sq_hinge_loss(arr):
  loss=[]
  for i in arr:
    loss.append(max(0,1-i)**2)
  return loss  
```

```python
z=np.linspace(-1,3,100000)
hin_loss=hinge_loss(z)
lg_loss=log_loss(z)
sq_hin_loss=sq_hinge_loss(z)
plt.plot(z,hin_loss,color='red',label='hinge_loss')
plt.plot(z,lg_loss,color='blue',label='log_loss')
plt.plot(z,sq_hin_loss,color='brown',label='squared_hinge_loss')
plt.xlabel('z values')
plt.ylabel('loss')
plt.legend()
plt.show()
```
