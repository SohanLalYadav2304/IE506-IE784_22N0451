### Ex2.Exercise 2: We would like to use words which are commonly used - let us assume that the wordle target word is picked at random from possibilities on the internet, and is therefore reflected in the word frequency. Assume that a list of such 5 letter words is available, in order of frequency. But we would like to avoid words in which a letter is used more than once. Can you filter out such words from the list given to you? C

```python
for i in range(26):
  df=df[df[letter_list[i]]<2]
df_non_repeating=df  
```

```python
df_non_repeating
```

```python
df_non_repeating.shape
```

### Since number of rows in df and df_non_repeating are same that means that in given list of 5 letter words, that is already filtered.

### Now we take some words with repeating letters and without repeating letters and then filter that.

```python
words=['again','algae','apart','arena','aroma','awake','award','basal','canal','carat','fauna','gamma','koala','madam','mamma','nasal','plaza','radar','salad','sauna']
```

```python
for i in range(20):
  words.append(df_non_repeating.iloc[i,0])
```

```python
len(words)
```

```python
list_words=pd.DataFrame(words,index=[i+1 for i in range(40) ],columns=['words'])
```

```python
list_words
```

```python
for j in range(26):
  list_words[letter_list[j]]=0
  for i in range(40):
    list_words.iloc[i,j+1]=list(list_words.iloc[i,0]).count(letter_list[j])
```

```python
list_words
```

```python
for i in range(26):
  list_words=list_words[list_words[letter_list[i]]<2]
```

```python
list_words
```

### Hence filtered list of words with non repeating letters.

```python
for i in range(list_words.shape[0]):
  print('',i+1,' ',list_words.iloc[i,0])
```

### Hence in the above method we can filter out words with non repeating letters form list of all words  of 5 letters.
