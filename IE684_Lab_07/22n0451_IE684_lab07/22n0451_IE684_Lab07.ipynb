```python
# Split Dataset into train and test dataset using sklearn
from sklearn.model_selection import train_test_split
X_data_train,X_data_test,Y_data_train,Y_data_test = train_test_split(X_data, Y_data, test_size=0.2, random_state=99)
len(X_data_train) , len(X_data_test) , len(Y_data_train) , len(Y_data_test)
```

```python

from sklearn.metrics import mean_squared_error,r2_score
```

```python
ratio=[0.4,0.3,0.1]
for i in ratio: 
  print('\n************************************************************\n')
  print('When train and test split:',100*(1-i),':',100*i)
  X_data_train,X_data_test,Y_data_train,Y_data_test = train_test_split(X_data, Y_data, test_size=i, random_state=99)
  len(X_data_train) , len(X_data_test) , len(Y_data_train) , len(Y_data_test)
  model=LinearRegression()
  model.fit(X_data_train,Y_data_train)
  print('coefficients:',model.coef_)
  print('intercept term',model.intercept_)
  Y_pred=model.predict(X_data_test)
  print('mean squared error in test data set is:',mean_squared_error(Y_data_test, Y_pred))
  print('R square is :',r2_score(Y_data_test,Y_pred))
```

```python
lambdas=[0.001,0.005,0.01,0.05,0.1,0.5]
```

```python
from sklearn.linear_model import Ridge,Lasso

for i in [0.2]+ratio: 
  print('\n************************************************************\n')
  print('When train and test split:',100*(1-i),':',100*i)
  X_data_train,X_data_test,Y_data_train,Y_data_test = train_test_split(X_data, Y_data, test_size=i, random_state=99)
  len(X_data_train) , len(X_data_test) , len(Y_data_train) , len(Y_data_test)
  for lembda in lambdas:
    print('___________________________________________')
    print('when lambda is :',lembda)
    print('Using Ridge Regression:\n')
    model=Ridge(alpha=lembda)
    model.fit(X_data_train,Y_data_train)
    print('coefficients:',model.coef_)
    print('intercept term:',model.intercept_)
    Y_pred=model.predict(X_data_test)
    print('mean squared error in test data set is:',mean_squared_error(Y_data_test, Y_pred))
    print('R square is :',r2_score(Y_data_test,Y_pred))
    print('\n Using Lasso Regression')
    model=Lasso(alpha=lembda)
    model.fit(X_data_train,Y_data_train)
    print('coefficients:',model.coef_)
    print('intercept term:',model.intercept_)
    Y_pred=model.predict(X_data_test)
    print('mean squared error in test data set is:',mean_squared_error(Y_data_test, Y_pred))
    print('R square is :',r2_score(Y_data_test,Y_pred))


```
